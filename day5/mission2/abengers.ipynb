## Stats Compare

```python
# attributes 추출
stats = [stat for stat in ab.axes[1][2:]]

# 시각화
fig, ax = plt.subplots(2, 3, figsize=(18, 10))

for s, a in enumerate(ax.ravel()):
    sns.kdeplot(ab[ab['Alignment'] == 'good'][stats[s]], ax=a)
    sns.kdeplot(ab[ab['Alignment'] == 'bad'][stats[s]], ax=a)
    sns.kdeplot(ab[ab['Alignment'] == 'neutral'][stats[s]], ax=a)
    
plt.legend(['good', 'bad', 'neutral'])
plt.show()
```

전반적으로 `neutral`성향의 캐릭터들이 강력한 능력치를 갖고 있는 것으로 판단된다.\
이는 자연 그 자체를 형상화한 캐릭터, 신, 우주적 존재 등 **천문학 레벨의 힘**을 가지면서도\
**선악의 구분이 모호한** 캐릭터들이 주로 `neutral`성향으로 설정되어있기 때문인 것으로 추정된다.
